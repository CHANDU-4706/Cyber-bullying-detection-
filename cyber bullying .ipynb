```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

import warnings
warnings.simplefilter("ignore")
```

```python
# Loading the dataset
data = pd.read_csv("bully_1.csv")
```

```python
data.head(10)
```

```python
# value count for each language
data["tagging"].value_counts()
```

```python
# separating the independent and dependant features
X = data["comments"]
y = data["tagging"]
```

```python
# converting categorical variables to numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

# Text Preprocessing

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
data_list = []
for text in X:
    text = re.sub(r'[!@#$(),.\n"%^*?\:;~`0-9]', ' ', text)
    text = re.sub(r'[[]]', ' ', text)
    text = re.sub(r'\[.*?\]', ' ', text)
    text = text.lower()
    data_list.append(text)

data = pd.DataFrame({'comments': data_list})
```

```python
data['Tokenized_Text'] = data['comments'].apply(word_tokenize)
data
```

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
data['Stemmed_Text'] = data['Tokenized_Text'].apply(lambda tokens: [porter_stemmer.stem(token) for token in tokens])
data
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
data['Lemmatized_Text'] = data['Tokenized_Text'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
data
```

```python
data['Char_Count'] = [len(text) for text in data['comments']]
data['Word_Count'] = [len(tokens) for tokens in data['Lemmatized_Text']]
data
```

```python
data_list1 = []
for text_list in data['Lemmatized_Text']:
    text = ' '.join(text_list)
    text = text.lower()
    data_list1.append(text)
```

```python
data_list1
```

```python
df=data.copy()
```

```python
from nltk.util import ngrams
text_data = df['Lemmatized_Text']
df['Bigrams'] = df['Lemmatized_Text'].apply(lambda x: list(ngrams(x, 2)))
df['Trigrams'] = df['Lemmatized_Text'].apply(lambda x: list(ngrams(x, 3)))
# df['Bigrams'] = df['Lemmatized_Text'].apply(lambda x: list(ngrams(x, 2)))
# df['Bigrams'] = df['Lemmatized_Text'].apply(lambda x: list(ngrams(x, 2)))
df
```

# Bag of Words

```python
# creating bag of words using countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(data_list1).toarray()
```

```python
print(X)
```

```python
X.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

# Initialize the QDA model
qda = QuadraticDiscriminantAnalysis()



```

```python
print(x_train)
print(y_train)
```

```python
# Fit the model to the training data
qda.fit(x_train, y_train)
```

```python
y_pred = qda.predict(x_test)
```

```python
qda.score(x_test, y_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Initialize the LDA model
lda = LinearDiscriminantAnalysis()

# Fit the model to the training data directly, without converting to array
lda.fit(x_train, y_train)
```

```python
y_pred = lda.predict(x_test)
```

```python
lda.score(x_test, y_test)
```

```python
import pickle

# save the iris classification model as a pickle file
model_pkl_file = "CyberBullyingQDA.pkl"  

with open(model_pkl_file, 'wb') as file:  
    pickle.dump(qda, file)
```
